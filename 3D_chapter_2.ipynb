```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

# Generate random point cloud (100 points with 3 coordinates)
point_clouds = np.random.rand(100, 3)

# Calculate mean of Z values
mean_Z = np.mean(point_clouds[:, 2])  # Direct calculation

# Spatial query: points where Z is within ±1 of mean_Z
mask = np.abs(point_clouds[:, 2] - mean_Z) < 1
spatial_query = point_clouds[mask]

print(f"Original points: {point_clouds.shape}")
print(f"Mean Z: {mean_Z:.4f}")
print(f"Filtered points: {spatial_query.shape}")

# Get XYZ coordinates
xyz = spatial_query[:, :3]  # All 3 columns are XYZ

# For random data, create synthetic RGB values (0-1 range)
rgb = np.random.rand(len(spatial_query), 3)  # Random colors

# Create 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(
    xyz[:, 0],  # X coordinates
    xyz[:, 1],  # Y coordinates
    xyz[:, 2],  # Z coordinates
    c=rgb,      # Colors (0-1 range, no division by 255 needed)
    s=20,       # Point size (increased from 0.01)
    alpha=0.7   # Transparency
)

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f'3D Point Cloud\n{len(spatial_query)} points within ±1 of mean Z')

plt.show()
```
